# Distrubtions and Plot Library 
This Module is used for making plots and testing the given column for various type of distributions like normal , welisberg , Exponential , Logistic etc . Find outliers in one column based on Z test 

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import distribution as db
a = pd.read_csv('sample data.csv')
a.head()
```

### hisplot1(dataset,name):
It is used to plot a histagram for a given dataset[name] 

```python
db.hisplot1(a,'Income')
```

### lineplot(dataset,name)
Used to do autoscaling and plot line for group of data "
Input: Name can be a list of features also 
Output:Line plot 

```python
db.lineplot(a,'Months Since Policy Inception')
```

### threed(dataset,name):
to draw a 3d plot
Input:Dataset : DataFrame and name is a list of 3 variables 
Output : 3d plot 

```python
db.threed(a,['Customer Lifetime Value','Income','Number of Open Complaints'])
```

### outliers1(dataset,name):
It is used to find the outliers in a numeric data series or a column in a dataframe based on Z test . It displays their Z score . It returns 1 if its an outlier and 0 if its not . 
Reference : https://onlinecourses.science.psu.edu/stat200/node/135

```python
db.outliers1(a,'Income') ## print the out
```

### typedis(dataset,name,dis):
Type any type of ditribution . Dis is used to take in the type of code distribution visit refer http://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html#module-scipy.stats for more reference and throws KS Test Statistic either D,D+,D-  test and p value as a result . we are taking p value as 0.05 to check a hypothesis with 95 % confidence 
Output : 
Zscore , Pvalue

```python
db.typedis(a,'Income','norm')
```

### norm(dataset,name):
Perform KS test for normal distribution .throws KS Test Statistic either D,D+,D-  test and p value as a result . we are taking p value as 0.05 to check a hypothesis with 95 % confidence 
Output : 
Zscore , Pvalue

```python
db.norm(a,'Income')
```
