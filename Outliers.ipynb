# Outlier Module 
This module contains functions to find outliers in a multivariate envirnment and to transform text data series to numeric one , To find outliers in a time series 

```python
import pandas as pd
import muloutlier as ml 
a = pd.read_csv('sample data.csv')
a.head()
```

### transform(dataset,col)
This function is used to map all text data series dataset[col] to numeric data . It can be passed any data series if numeric it won't be altered . 
Output: An altered data frame 

```python
for i in a:
    a = ml.transform(a,i)
a.head()    
```

### mutout(df)
Here df is the data frame . It considers all the columns of a data frame in both numeric and text form . It performs automatic transformation and preprocessing as well. It then uses the concept of oneSVM using rbf kernal evaluate the records which can be outliers .
Output: All the record numbers which are outliers and throws back a list which contains -1  to that a given value is an outlier and +1 to indicate that its not an outlier

```python
ml.mulout(a)
```

## Outliers for a time series  

```python
b = pd.read_csv('time series sample.csv')
b.head()
```

### timeoutlier(dataset, col, window):
It is used to find outliers for time series .Here window is the window size for which an outlier should be calculated . It also draws a plot showing the the ranges and data points for a time series 
Output :
list with a value of True if an outlier and false if not an outlier

```python
ml.timeoutlier(b,'High',3)
```
