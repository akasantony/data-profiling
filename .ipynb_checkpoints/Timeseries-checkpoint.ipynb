# Time Series
This module is used to do analysis of time series . 

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
a = pd.read_csv('time series sample.csv')
a.head()
```

```python
a['Open'].plot(x=a['Date'].astype(str))
plt.show()
```

### Rollingmean(dataframe,col,window)
It is used to calculate the rolling mean for dataframe[col] . Here Window is the size of the window 
Output: It returns a data series of rolling mean of window size 

```python
import split as sp
k = sp.rollingmean(a,'Open',3)
k.plot(x=a['Date'].astype(str))
plt.show()
```

### rollingstd(dataframe,col,window)
It is used to calculate the rolling standard deviation  for dataframe[col] . Here Window is the size of the window 
Output: It returns a data series of rolling mean of window size 

```python
k = sp.rollingstd(a,'Open',3)
k.plot(x=a['Date'].astype(str))
plt.show()
```

### timeoutlier(dataset, col, window):
It is used to find outliers for time series .Here window is the window size for which an outlier should be calculated . It also draws a plot showing the the ranges and data points for a time series 
Output :
list with a value of True if an outlier and false if not an outlier

```python
import muloutlier as ml 
ml.timeoutlier(a,'High',7)
```

## Decomposition Analysis

Decomposition analysis essentially allows us to separate out the components of each time series. Separating the components out helps us understand the impact of trends, seasonality and random noise on the time series data. It also allows us to make predictions on the series.

## decompose(df,col,freq):
To plot the decomposition graphs of time series of col in df data frame of freq frequency . 
Output:
Observed 
Trend
Seasonal 
Residual Graphs 

```python
import timeseries as ts
ts.decompose(a,'Open',10)
```

### check(df,col):
To check whether a given series is Periodic or not using AutoCorrelation function . Plot ACF With lags 

```python
ts.check(a,'Open')
```

## LM Test 
H0: The data are independently distributed (i.e. the correlations in the population from which the sample is taken are 0, so that any observed correlations in the data result from randomness of the sampling process).
Ha: The data are not independently distributed; they exhibit serial correlation.

### lmtestcheck(df,col,max1):
To perform and LM test for autocorrelation and find significant lags . 
Output : 1 to determine a significant lag and 0 to determine insignificant lag

```python
ts.lmtestcheck(a,'Open',40)
```

## checkdb(df,col):
It also tells whether the Data is serially correlated or not 

```python
ts.checkdb(a,'Open')
```

### freq(df,col,max1):
To find the required freq for the decompostion on which data is viable or not . max1 is the maximum frequency for which you want to test . 
Output : Frequency on which Decompostion is viable or not 

```python
ts.freq(a,'Open',a['Open'].count())
```
